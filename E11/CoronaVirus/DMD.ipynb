```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
country_choice = ['Norway','Italy','Iran','India','Germany','United Kingdom','US']
country_choice = ['Norway']
```

```python
confirmed = pd.read_csv('./data/time_series_covid_19_confirmed.csv',sep=',')
recovered = pd.read_csv('./data/time_series_covid_19_recovered.csv',sep=',')
deaths = pd.read_csv('./data/time_series_covid_19_deaths.csv',sep=',')
```

```python
data=confirmed
```

```python
data(['deaths'])=deaths
```

```python
pd.DataFrame.transpose(data)
```

```python
data2=data.groupby('Country/Region').sum()
data2=data2.drop(['Lat','Long'],axis=1)
data2=pd.DataFrame.transpose(data2)
data2[country_choice].plot(figsize=(15,10),logy=True)
plt.grid()
```

```python
X=data2.values
```

```python
# Confirmed
fig = px.choropleth(data2[10:11], locations="Country/Region", 
                    locationmode='country names', color="Confirmed", 
                    hover_name="Country/Region", range_color=[1,7000], 
                    color_continuous_scale="aggrnyl", 
                    title='Countries with Confirmed Cases')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
