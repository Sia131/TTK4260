# Importing required libraries

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

# Reading the data

```python
# This dataset doesn't have headers to describe the data, 
# therefore to load it corretly into a pandas dataframe we spcify "header" and "inxed_col" parmeters as shown below:
data = pd.read_csv('../data/Microchip/microchip.csv',header=None, index_col=False)
```

# Getting an overview of the data

```python
data.head()
```

### Data size

```python
print(f"Size (row, col): {data.shape}")
print(f"Unique values in 3rd col: {data[2].unique()} --> categorical variable" )
```

### Plotting data

```python
## Plots
fig,axs = plt.subplots(2,1,figsize=(15,10))
fig.tight_layout(pad=3.0)

plt.sca(axs[0])
plt.plot(data[0], c='r', label="0")
plt.plot(data[1], c='g', label="1")
# plt.title('')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid()
plt.legend(loc='upper right')

plt.sca(axs[1])
plt.plot(data[2], c='b', label="3")
#plt.title(')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid()
plt.legend()

fig.suptitle('Line plot Microchip data', fontsize=20)
plt.show()
```
